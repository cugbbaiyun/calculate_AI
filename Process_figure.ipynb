```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def iterName(i, layer, func):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
def makefig(i,layer=1, func='vary_sgd_'):
   # sns.color_palette('muted')
    file_name, result_name = iterName(i, layer, func)
    datas = pd.read_csv(file_name, sep=',')
    print(result_name)
    resultdatas = pd.read_csv(result_name)
    print(resultdatas)
    pd1 = datas[['iter', 'train_accu']]
    pd2 = datas[['iter', 'test_accu']]
    pd3 = resultdatas[['iter', 'result_accu']]
    pd1.columns = ['iter', 'accu']
    pd2.columns = ['iter', 'accu']
    pd3.columns = ['iter', 'accu']
    pd1['type'] = pd.Series(['train' for _ in range(0, len(pd1))])
    pd2['type'] = pd.Series(['test' for _ in range(0, len(pd2))])
    pd3['type'] = pd.Series(['result' for _ in range(0,len(pd3))])
    #print(pd3.tail(10))
    pd4 = pd.concat([pd1, pd2, pd3])
    ax = sns.lineplot(x='iter', y='accu', hue='type', data=pd4, palette='deep')
    plt.savefig("C:\\Users\\win10\\Desktop\\calaiwork\\vary_sgd\\"+str(layer) + "_"+ str(i) +"\\ans.png")
    plt.show()
```

```python
for i in range(1, 5):
    for j in range(1, 10):
        makefig(j, i)
```
