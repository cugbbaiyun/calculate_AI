```python
def iterName(i, layer, func):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    #result_name += '-'
    #result_name += str(gamma)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
import numpy as np
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
func = ['SGD_', 'SGD_sigmoid_', 'SGD_sigmoid_tanH_']
```

```python
layer = [1]
node = [1,3,5]
```

```python
def makefig(layer, node):
    for n in node:
        _, filn1=iterName(n, layer, func[0])
        _, filn2=iterName(n, layer, func[1])
        _, filn3=iterName(n, layer, func[2])
        data1 = pd.read_csv(filn1)
        data2 = pd.read_csv(filn2)
        data3 = pd.read_csv(filn3)
        data1['activ'] = pd.Series(['tanh' for _ in range(len(data1))])
        data2['activ'] = pd.Series(['sigmoid' for _ in range(len(data2))])
        data3['activ'] = pd.Series(['sigmoid+tanh' for _ in range(len(data3))])
        print(data1.head(10))
        print(data2.head(10))
        data = pd.concat([data1,data2, data3])
        ax = sns.lineplot(data=data, x='iter', y='result_accu', hue='activ')
        plt.show()
```

```python
makefig(1, node)
```
