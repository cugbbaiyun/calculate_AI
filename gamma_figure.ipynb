```python
def iterName(i, layer, func,gamma):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    result_name += '_'
    result_name += str(gamma)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
import numpy as np
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
def makefig(gammas, layer_, node_num):
    _, resn1 = iterName(node_num, layer_, 'vary_sgd_', gammas[0])
    _, resn2 = iterName(node_num, layer_, 'vary_sgd_', gammas[1])
    _, resn3 = iterName(node_num, layer_, 'vary_sgd_', gammas[2])
    _, resn4 = iterName(node_num, layer_, 'vary_sgd_', gammas[3])
    data1 = pd.read_csv(resn1)
    data2 = pd.read_csv(resn2)
    data3 = pd.read_csv(resn3)
    data4 = pd.read_csv(resn4)
    data1['gamma'] = pd.Series(['gamma=' + str(gammas[0]) for _ in range(len(data1))])
    data2['gamma'] = pd.Series(['gamma=' + str(gammas[1]) for _ in range(len(data2))])
    data3['gamma'] = pd.Series(['gamma=' + str(gammas[2]) for _ in range(len(data3))])
    data4['gamma'] = pd.Series(['gamma=' + str(gammas[3]) for _ in range(len(data4))])
    data = pd.concat([data1,data2,data3,data4])
    ax = sns.lineplot(data=data, x='iter', y='result_accu', hue='gamma')
    plt.show()
```

```python
gamma=['0.03','0.09','0.1','0.15']
layer =[1,2,3,4]
node_num = 5
for layer_ in layer:
    makefig(gamma, layer_, node_num)
```
