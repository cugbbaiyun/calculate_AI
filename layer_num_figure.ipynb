```python
def iterName(i, layer, func):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
import numpy as np
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
layer = [1,2,3,4]
num = [1,5,9]
func=['SGD_', 'vary_sgd_', 'momentum_']
```

```python
def makefig(func, layer, num):
    data1 = None
    data2 = None
    data3 = None
    data4 = None
    for num_ in num:
        _, result_name1 = iterName(num_, layer[0], func)
        _, result_name2 = iterName(num_, layer[1], func)
        _, result_name3 = iterName(num_, layer[2], func)
        _, result_name4 = iterName(num_, layer[3], func)
        data1 = pd.read_csv(result_name1)
        data2 = pd.read_csv(result_name2)
        data3 = pd.read_csv(result_name3)
        data4 = pd.read_csv(result_name4)
        data1['layer_num'] = pd.Series(['layer_num=1' for _ in range(len(data1))])
        data2['layer_num'] = pd.Series(['layer_num=2' for _ in range(len(data2))])
        data3['layer_num'] = pd.Series(['layer_num=3' for _ in range(len(data3))])
        data4['layer_num'] = pd.Series(['layer_num=4' for _ in range(len(data4))])
        data = pd.concat([data1, data2, data3, data4])
        ax = sns.lineplot(data=data, x ='iter', y='result_accu', hue='layer_num')
        plt.show()
```

```python
makefig(func[0], layer, num)
```

```python
makefig(func[1], layer, num)
```
