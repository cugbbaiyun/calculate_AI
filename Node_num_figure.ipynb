```python
import numpy as np
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
def iterName(i, layer, func):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
def makefig2(rg, layer):
    _, result_name1 = iterName(rg[0],layer,'vary_sgd_')
    data1 = pd.read_csv(result_name1, sep=',')
    _, result_name2 = iterName(rg[1],layer,'vary_sgd_')
    data2 = pd.read_csv(result_name2, sep=',')
    _, result_name3 = iterName(rg[2],layer,'vary_sgd_')
    data3 = pd.read_csv(result_name3, sep=',')
    data1['num'] = pd.Series(['num=' + str(rg[0]) for _ in range(len(data1))])
    data2['num'] = pd.Series(['num=' + str(rg[1]) for _ in range(len(data2))])
    data3['num'] = pd.Series(['num=' + str(rg[2]) for _ in range(len(data3))])
    data = pd.concat([data1, data2, data3])
    ax = sns.lineplot(data=data, x ='iter', y='result_accu', hue='num')
    plt.show()
```

```python
def makefig(layer):
    rg = [[1,2,3], [3,4,5],[5,6,7],[7,8,9]]
    for c in rg:
        makefig2(c , layer)
```
