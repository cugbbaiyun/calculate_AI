```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def iterName(i, layer, func):
    file_name = func + 'process_'  + str(layer)
    result_name = func + 'result_' + str(layer)
    for j in range(layer):
        file_name+= "_" + str(i)
        result_name += "_" + str(i)
    file_name += ".log"
    result_name += ".log"
    return file_name, result_name
```

```python
layer_node = [(1,1), (1,5),(1,9), (2,1), (2,5), (2,9),(3,1),(3,5),(3,9)]
funcs = ['SGD_', 'vary_sgd_', 'momentum_']
```

```python
def makefig(layer, node):
    _, result_name1 = iterName(node, layer, funcs[0])
    _, result_name2 = iterName(node, layer, funcs[1])
    _, result_name3 = iterName(node, layer, funcs[2])
    #_, result_name4 = iterName(node, layer, funcs[3])
    data1 = pd.read_csv(result_name1)
    data2 = pd.read_csv(result_name2)
    data3 = pd.read_csv(result_name3)
    #data4 = pd.read_csv(result_name4)
    print(data1.head(10))
    print(data2.head(10))
    print(data3.head(10))
    data1['optim'] = pd.Series(['sgd' for _ in range(len(data1))])
    data2['optim'] = pd.Series(['vary_sgd' for _ in range(len(data2))])
    data3['optim'] = pd.Series(['momentum_sgd' for _ in range(len(data3))])
    #data4['optim'] = pd.Series(['momentum' for _ in range(len(data4))])
    data = pd.concat([data1,data2,data3])
    ax = sns.lineplot(data=data,x = 'iter', y='result_accu',hue ='optim')
    plt.show()
```

```python
for layer,node in layer_node:
    makefig(layer, node)
```
